```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

import plotly.graph_objs as go

print(__version__) 

init_notebook_mode(connected=True)
```

```python
trace1 = go.Bar(
    y = portfolio_df['house_id'][0:2000],
    x = portfolio_df['house_year'][0:2000],
    name = 'Количество зданий, построенных в этот год')

trace2 = go.Bar(
    y = portfolio_df['area_id'][0:2000],
    x = portfolio_df['house_year'][0:2000],
    name = 'Количество зданий, построенных в этот год')

data = [trace1, trace2]

layout = go.Layout(title = 'Количество зданий, построенных в этот год'
    , barmode = 'group'
    , xaxis=dict(title='Год постройки')
    , yaxis=dict(title='Количество зданий')
    , legend=dict(x=0.8,y=1.2)
    )

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
trace1 = go.Bar(
    y = portfolio_df['area_name'][0:2000],
    x = portfolio_df['area_id'][0:2000],
    name = 'Количество зданий в этом МО')

trace2 = go.Bar(
    y = portfolio_df['area_name'][0:2000],
    x = portfolio_df['house_id'][0:2000],
    name = 'Количество зданий в этом МО')

data = [trace1, trace2]

layout = go.Layout(title = 'Количество зданий в этом МО'
    , barmode = 'group'
    , xaxis=dict(title='Количество зданий')
    , yaxis=dict(title='Муниципальный округ')
    , legend=dict(x=0.8,y=1.2)
    )

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
