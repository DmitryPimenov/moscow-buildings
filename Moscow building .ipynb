```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

import plotly.graph_objs as go

print(__version__) 

init_notebook_mode(connected=True)
```

1. Построить график, на котором по оси абсцисс – год постройки, по оси ординат – количество зданий, построенных в этот год.

```python
trace1 = go.Bar(
    y = portfolio_df['house_id'][0:29464],
    x = portfolio_df['house_year'][0:2019],
    name = 'Количество зданий, построенных в этот год')

data = [trace1]

layout = go.Layout(title = 'Количество зданий, построенных в этот год'
    , barmode = 'group'
    , xaxis=dict(title='Год постройки')
    , yaxis=dict(title='Количество зданий')
    , legend=dict(x=0.8,y=1.2)
    )

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

2) Простая задача, можно решить любым способом. Необходимо узнать является ли слово палиндромом без учета регистра (ШаЛАш - палиндром) и без учета пробелов (Ша Ла ш - тоже палиндром). На вход подается строка, на выходе True или False.

```python
word = input().lower()

def palindrom(word):
    if word==word[::-1]:
        print(word, True)
    else:
        print(word, False)
        

palindrom(word)


a, b, c = input().split(' ')
word = a + b + c

palindrom(word.lower())
```

4) Необходимо заполнить признак "Отопление" (сейчас его нет). Он может принимать два значения - "Да" и "Нет". Необходимо вытащить его из Описания. Если Отопление есть в описании и не указано, что его нет, то считаем, что отопление есть (т.е. например "центральное отопление", "отопление есть" и т.д. = "Да", а если явно указано что его нет, то = "Нет"). Не обязательно все строки должны быть заполнены, где-то может и не быть отопления. В этом задании мы хотим посмотреть как Вы умеете работать с данными и регулярными выражениями.

```python
import re
desc = df['Description'].str.split(n=1).str[1]

for val in desc:
    if re.findall(r'отапливаемое помещение', val) or re.findall(r'отопление', val) or re.findall(r'отоплением', val):
        print ('Да')
    else:
        print ('Нет')
```

5) Построить график, на котором по оси абсцисс – класс и тип сделки, по оси ординат – цена сделки.

```python
trace1 = go.Bar(
    y = df['Price'][0:100],
    x = df['Class'][0:100],
    name = 'Цена по классам сделок')

trace2 = go.Bar(
    y = df['Price'][0:100],
    x = df['Dealtype'][0:100],
    name = 'Цена по типу сделки')

data = [trace1, trace2]

layout = go.Layout(title = 'Цена'
    , barmode = 'group'
    , xaxis=dict(title='Класс и тип сделки')
    , yaxis=dict(title='Цена сделки')
    , legend=dict(x=0.8,y=1.2)
    )

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
